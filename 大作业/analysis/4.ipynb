```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.sans-serif'] = ['SimHei']

data = pd.read_csv('../datasets/all.csv')

statistics = data.groupby(['city', 'room_type']).agg({
    'price': ['mean', 'max', 'min', 'median']
}).reset_index()

filtered_data = statistics[statistics['room_type'].isin(['1', '2', '3'])]
print(filtered_data)
```

```python
# 获取不同 room_type 的数据
room_type_1 = filtered_data[filtered_data['room_type'] == '1']
room_type_2 = filtered_data[filtered_data['room_type'] == '2']
room_type_3 = filtered_data[filtered_data['room_type'] == '3']

print(room_type_1)
```

```python
room_type_1.plot(kind='bar', x='city', y='price', legend=True,
                 title='一居情况', xlabel='城市', ylabel='价格')
plt.yscale('log')
plt.savefig('imgs/一居情况.png')

room_type_2.plot(kind='bar', x='city', y='price', legend=True,
                 title='二居情况', xlabel='城市', ylabel='价格')
plt.yscale('log')
plt.savefig('imgs/二居情况.png')

room_type_3.plot(kind='bar', x='city', y='price', legend=True,
                 title='三居情况', xlabel='城市', ylabel='价格')
plt.yscale('log')
plt.savefig('imgs/三居情况.png')
```
